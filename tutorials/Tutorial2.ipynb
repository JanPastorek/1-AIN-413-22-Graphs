# Havel-Hakimi recursive algorithm to find whether a seq is graphical

```python
import networkx as nx
import random
```

```python
# generate a random graph
G = nx.gnp_random_graph(10, 0.5)

# print the degree sequence
print("Degree sequence", sorted([d for n, d in G.degree()], reverse=True))
```

```python
# generate a random sequence of numbers
seq = [random.randint(0, 8) for i in range(10)]
print("Random sequence", seq)

# find out if the sequence is a degree sequence
print("Is this sequence of numbers a graphical sequence?", nx.is_graphical(seq, method="hh"))

# if it is a graphical sequence, create a graph from it
if nx.is_graphical(seq):
    
    # create a simple graph from the sequence using the Havel-Hakimi algorithm
    H = nx.havel_hakimi_graph(seq)
    
    # plot the graph
    nx.draw(H, with_labels=True)
```
